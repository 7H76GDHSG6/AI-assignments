#### We now group similar roles into 1 common role

```python
df['job role'].replace(['Design & UX', 'UX Designer'], 'UX & Design', inplace=True)
df['job role'].replace(['Business Systems Analyst', 'Business Intelligence Analyst', 'Programmer Analyst',
                        'CRM Business Analyst', 'Systems Analyst', 'Information Security Analyst', 'E-Commerce Analyst'],
                       'Analyst', inplace=True)
df['job role'].replace(['Database Manager', 'Information Technology Manager', 'Project Manager'], 'Manager', inplace=True)
df['job role'].replace(['Network Security Administrator', 'Database Administrator', 'Portal Administrator',
                        'Systems Security Administrator'], 'Administrator', inplace=True)
df['job role'].replace(['Software Developer', 'Database Developer', 'Web Developer',
                        'CRM Technical Developer', 'Applications Developer', 'Mobile Applications Developer'],
                       'Developer', inplace=True)
df['job role'].replace(['Network Security Engineer', 'Network Engineer', 'Software Engineer',
                        'Software Systems Engineer', 'Technical Engineer'], 'Engineer', inplace=True)
df['job role'].replace(['Technical Support', 'Technical Services/Help Desk/Tech Support'], 'Support', inplace=True)
df['job role'].replace(['Solutions Architect', 'Data Architect'], 'Architect', inplace=True)
df['job role'].replace(['Quality Assurance Associate', 'Information Technology Auditor',
                        'Software Quality Assurance (QA) / Testing'], 'Quality Assurance', inplace=True)
```

```python
df = pd.get_dummies(prefix='cert', data=df, columns=['certifications'])
df = pd.get_dummies(prefix='wshop', data=df, columns=['workshops'])
df = pd.get_dummies(prefix='subj', data=df, columns=['interest subject'])
df = pd.get_dummies(prefix='career', data=df, columns=['interest career'])
df = pd.get_dummies(prefix='comp', data=df, columns=['company choice'])
```

```python
df.shape
```

## Defining the model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def make_model(dim = df.shape[1] - 1):
    model = Sequential()
    model.add(Dense(units=32, input_dim = dim, activation='tanh', kernel_initializer='he_uniform'))
    model.add(Dense(units=64, activation='tanh', kernel_initializer='he_uniform'))
    model.add(Dense(units=16, activation='tanh', kernel_initializer='he_uniform'))
    model.add(Dense(units=9, activation='softmax', kernel_initializer='he_uniform'))
    model.summary()
    model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
    return model
```

```python
model = make_model()
```

## Training and testing

```python
x = df.drop('job role', axis=1)
y = df['job role']
```

```python
def train(model, x_train, y_train, bsize=256, nepoch=64):
    for i in range(4):
        model.fit(x_train, y_train, batch_size = bsize, epochs = nepoch,verbose = 1)
        bsize //= 2
        nepoch //= 2
```

### 70:30 split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
model = make_model()
```

```python
train(model, x_train, y_train)
```

```python
score, acc = model.evaluate(x_train, y_train, batch_size=100)
```

```python
preds = model.predict(x_test)
preds = np.argmax(preds, axis=1)

score, acc = model.evaluate(x_test, y_test, batch_size=100)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
targets = job_encoder.inverse_transform([0, 1, 2, 3, 4, 5, 6, 7, 8])
print(classification_report(y_test, preds, target_names = targets))
```

```python
cmat = confusion_matrix(y_test, preds)
plt.figure(figsize = (12,8))
sns.heatmap(cmat, annot=True, fmt='d')
```

### class-wise accuracy
